```python
import numpy as np
import pandas as pd

# Color palette for visualizations
palette = "viridis"

# Read dataset
filename = "heart.csv"
df = pd.read_csv(filename)
df.head()
```

```python
corr = df.corr()
corr.style.background_gradient(cmap=palette)
```

```python
import utils
dff = df
dff["age"] = utils.normalize(df["age"])
dff["cp"] = utils.normalize(df["cp"])
dff["trestbps"] = utils.standardize_normal(df["trestbps"])
dff["chol"] = utils.standardize_normal(df["chol"])
dff["thalach"] = utils.standardize_normal(df["thalach"])
dff["oldpeak"] = utils.normalize(df["oldpeak"])
dff["slope"] = utils.normalize(df["slope"])
dff["thal"] = utils.normalize(df["thal"])
dff["ca"] = utils.normalize(df["ca"])


import matplotlib.pyplot as plt

# Plot the target variable
plt.title("TrestBPS")
plt.hist(df["trestbps"])
plt.show()
plt.title("chol")
plt.hist(df["chol"])
plt.show()
plt.title("thalach")
plt.hist(df["thalach"])
plt.show()
plt.title("oldpeak")
plt.hist(df["oldpeak"])
plt.show()
```

```python
dff = dff.sample(frac=1)
df_split = np.array_split(dff, 2)

training = df_split[0]
training_labels = training.iloc[:, -1].values
training_inputs = training.iloc[:, :-1].values
print(training_inputs)
print(training_labels)
```
